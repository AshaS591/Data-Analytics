```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/winequalityN.csv')
```

```python
df.columns
```

# Heat Map

```python
var=df.pivot_table(values=['alcohol','pH'],index='type')
```

```python
sns.heatmap(var,annot=True,cmap='rainbow')
```

```python
var=df.iloc[:,1:].corr()
sns.heatmap(var,annot=True,cmap='coolwarm')
```

```python
var=df.iloc[:,1:4].corr()
sns.heatmap(var,annot=True,cmap='coolwarm')
```

```python
var=df.iloc[1:,5:8].corr()
sns.heatmap(var,annot=True,cmap='magma')
```

```python
var=df.iloc[:,2:8].corr()
sns.heatmap(var,annot=True,cmap='rainbow')
```

```python
import matplotlib.pyplot as plt
# var=plt.figure(figsize=(10,20))
var=df.pivot_table(values=['pH','alcohol'],index='quality')
sns.heatmap(var,annot=True,cmap='coolwarm')
```

```python
var=df.pivot_table(values=['quality','alcohol','sulphates','free sulfur dioxide'],index='type')
sns.heatmap(var,annot=True,cmap='coolwarm')
```
