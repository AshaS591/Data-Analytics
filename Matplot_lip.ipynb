```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,7)
y=x**2
plt.plot(x,y)
```

```python
x=np.arange(1,7)
y=x**2
plt.plot(x,y,color='deeppink')
x1=np.arange(1,7)
y1=x**3
z=x*2
plt.plot(x1,y1,z,color='navy')
```

```python
x=np.linspace(1,7,3)
y=x**2
plt.plot(x,y,color='deeppink')
x1=np.linspace(1,7,3)
y1=x**3

plt.plot(x1,y1,color='navy')
```

```python
x=np.arange(1,7)
y=x**2
z=x**3
plt.plot(x,z,color='navy',label='cube',marker='o',markersize='8',markerfacecolor='yellow',markeredgecolor='red',linestyle='dashed')
plt.xlabel('values from 1 to 7',fontsize=15,color='blue')
plt.ylabel('squares from 1 to 7',fontsize=15,color='orange')
plt.title('Linear Graph',color='green',fontsize=20,fontstyle='italic')
plt.plot(x,y,color='deeppink',label='square',marker='*',markersize='15',markerfacecolor='green',markeredgecolor='orange',linestyle='-.')
plt.legend()
plt.show() # removes unwanted adderesses and avoids rendering
```

```python
x=np.arange(1,7)
y=x**2
z=x**3
plt.plot(x,z,color='navy',label='cube',marker='o',markersize='8',markerfacecolor='yellow',markeredgecolor='red',linestyle='dashed',linewidth='3')
plt.xlabel('values from 1 to 7',fontsize=15,color='blue')
plt.ylabel('squares from 1 to 7',fontsize=15,color='orange')
plt.title('Linear Graph',color='green',fontsize=20,fontstyle='italic')
plt.plot(x,y,color='deeppink',label='square',marker='*',markersize='15',markerfacecolor='green',markeredgecolor='orange',linestyle='-.',linewidth='3')
plt.legend()
plt.show() # removes unwanted adderesses and avoids rendering
```

```python
x=np.arange(4,10)
y=x+2
plt.plot(x,y,color='red',marker='*',markerfacecolor='navy',markeredgecolor='yellow',markersize=15,linestyle='-.',linewidth=3,label='4 to 9')
plt.xlabel('values from 4 to 9')
plt.ylabel('values from 6 to 11')
plt.title('Linear Graph')
plt.legend()
plt.show()
```

```python
x=np.arange(4,10)
y=[1,2.5,8,9.8,10,2]
plt.plot(x,y,color='red',marker='*',markerfacecolor='navy',markeredgecolor='yellow',markersize=15,linestyle='-',linewidth=3,label='4 to 9')
plt.xlabel('values from 4 to 9',color='purple')
plt.ylabel('values from 6 to 11',color='navy')
plt.title('Linear Graph',color='deeppink',fontstyle='italic')
plt.legend()
plt.show()
```

```python
x=[num for num in range(1,10,2)]
y=[num for num in range(0,10,2)]
x1=[num for num in range(0,10)]
z=[num for num in range(10,0,-1)]
plt.plot(x,y,marker='d',markerfacecolor='red',markersize=15,markeredgecolor='black',label='Range',linewidth=5)
plt.plot(x1,z,marker='*',markersize=20,markerfacecolor='green',markeredgecolor='black',label='opposite line',linewidth=5)
plt.title('Practice')
plt.legend()
plt.show()
```

```python
x=[num for num in range(1,10,2)]
y=[num for num in range(0,10,2)]
x1=[num+2 for num in range(0,10)]
z=[num-2 for num in range(10,0,-1)]
plt.plot(x,y,marker='d',markerfacecolor='red',markersize=15,markeredgecolor='white',label='Range',linewidth=5)
plt.plot(x1,z,marker='*',markersize=20,markerfacecolor='green',markeredgecolor='orange',label='opposite line',linewidth=5)

plt.legend()
plt.show()
```
