```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,7)
y=x**2+2
z=y+x
p=y+z
q=x+p
```

```python
plt.subplot(3,2,1)
plt.plot(x,y,color='red',label='Numbers Graph')
plt.plot(p,z,color='darkgreen')
plt.xlabel('no. from 1 to 5')
plt.ylabel('num square +2')
plt.legend()

plt.subplot(3,2,2)
# plt.plot(x,z,color='navy')
# plt.xlabel('no. from 1 to 5')
# plt.ylabel('num square +2')

plt.subplot(3,2,3)
plt.plot(x,p,color='deeppink')
plt.xlabel('no. from 1 to 5')
plt.ylabel('num square +2')

plt.subplot(3,2,4)
plt.plot(z,y,'blue')
plt.xlabel('no. from 1 to 5')
plt.ylabel('num square +2')

plt.subplot(3,2,5)
plt.plot(p,y,color='black')

plt.subplot(3,2,6)
plt.plot(p,q,color='green')

plt.tight_layout()
```
