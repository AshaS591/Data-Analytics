```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({'Ename':['Asha','Anu','Arya','Amulya','shashi'],
                 'Eid':[1,2,3,4,5],
                  'sal':[10000,2000,3000,9000,30000],
                 'Deptno':[1,4,1,2,8]})
df1
```

```python
df2=pd.DataFrame({'Dname':['Dev','TSG','Devops','IT'],
                 'Deptno':[1,2,3,4],
                 'loc':['KA','TN','BR','MP']})
df2
```

```python
df3=pd.merge(df1,df2,on='Deptno',how='inner')
df3
```

```python
df4=pd.merge(df1,df2,on='Deptno',how='left')
df4
```

```python
df5=pd.merge(df1,df2,on='Deptno',how='right')
df5
```

```python
df6=pd.merge(df1,df2,on='Deptno',how='outer')
df6
```
