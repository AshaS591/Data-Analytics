#**Data** **visualisation**

```python
#Based on 'EthnicGroup'
factor=df.pivot_table(values=['MathScore','ReadingScore','WritingScore'],index='EthnicGroup')
sns.heatmap(factor,annot=True)
plt.show()
```

**Students from group E and group D scored more compared to others**

```python
#Based on'ParentEduc'
factor=df.pivot_table(values=['MathScore','ReadingScore','WritingScore'],index='ParentEduc')
sns.heatmap(factor,annot=True)
plt.show()
```

**parents who have master degree, bachelor's degree and associate degree ,their children scored more compared to others**

```python
#Based on 'LunchType'
factor=df.pivot_table(values=['MathScore','ReadingScore','WritingScore'],index='LunchType')
sns.heatmap(factor,annot=True)
plt.show()
```

**students who having standard lunch type they scored more compared to others**

```python
#Based on 'TestPrep'
factor=df.pivot_table(values=['MathScore','ReadingScore','WritingScore'],index='TestPrep')
sns.heatmap(factor,annot=True)
plt.show()
```
