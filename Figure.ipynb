```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
var=plt.figure(figsize=(9,10),facecolor='yellow')
a1=var.add_axes([0.1,0.1,0.8,0.8])
a2=var.add_axes([0.5,0.5,0.4,0.4])

x=np.arange(1,6)
y=x**2
z=x**3

a1.plot(x,y,color='navy',linestyle='-.',label='Square')
a1.set_xlabel('val from 1 to 5')
a1.set_ylabel('Sqr from 1 to 5')
a1.set_title('Squares')
a1.set_facecolor('pink')
a1.legend()


a2.plot(y,z,color='crimson',linestyle=':',label='Cube')
a2.set_xlabel('val from 1 to 5')
a2.set_ylabel('cube from 1 to 5')
a2.set_title('Cubes')
a2.set_facecolor('skyblue')

a2.legend()
```
