```python
q1=df1['alcohol'].quantile(0.25)
q3=df1['alcohol'].quantile(0.75)

IQR=q3-q1

lower=q1-1*IQR
upper=q3+1*IQR

out=df1[(df1['alcohol']>upper) | (df1['alcohol']<lower) ]

df2=df1[(df1['alcohol']<upper) & (df1['alcohol']>lower) ]
median=df2['alcohol'].median()
median
a=out.index
for index in a:
  df1['alcohol'][index]=median
```
