```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/amazon_fires.csv',encoding='ISO-8859-1')
df
```

```python
df['year']=[year.split('/')[2] for year in df['encontro']]
df
```

```python
df['day']=[year.split('/')[0] for year in df['encontro']]
df
```

```python
df=pd.DataFrame({'NAME':['A','B','C','D','E','F','G','H'],
                 'ROLL':[101,102,103,104,105,106,107,108],
                 'KANNADA':[99,89,95,80,75,76,90,92],
                 'HINDI':[92,85,78,90,84,91,76,87],
                 'ENGLISH':[90,89,70,92,89,90,86,74],
                 'MATH':[99,80,95,78,83,86,84,70],
                 'SCIENCE':[90,87,67,89,92,80,76,87],
                 'SS':[89,78,70,75,87,73,70,87]
                 })
df
```

```python
df['PERCENTAGE']=round((df['KANNADA']+df['HINDI']+df['ENGLISH']+df['MATH']+df['SCIENCE']+df['SS'])/6)
df
```

```python
res=[]
for per in df['PERCENTAGE']:
  if 100>=per>=90:
    res.append('FCD')
  elif per>=80:
     res.append('FC')
  elif per>=60:
    res.append('SC')
  elif per>=35:
    res.append('JS')
  elif per<35:
    res.append('Fail')

print(res)
```

```python
df['RESULT']=res
df
```
