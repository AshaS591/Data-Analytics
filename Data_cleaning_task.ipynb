```python
import pandas as pd
```

```python
df=pd.read_csv('/content/employee_data.csv')
```

```python
df
```

#Rename Columns:
#The column names are not standardized.
##Rename the following columns:
###▪ full_name to name
###▪ emp_id to employee_id
###▪ emp_status to employment_status

```python
d={'full_name':'name','emp_id':'employee_id','emp_status':'employment_status'}
df.rename(columns=d,inplace=True)
df
```

#2. Rearrange Columns:

##Rearrange the columns in this order:
 ▪ employee_id, name, age, gender, department,join_date,
exp_years,salary,employment_status

```python
df.info()
```

```python
items=[0,1,2,3,4,6,7,5,8]
df=df[df.columns[items]]
df
```

#3. Update Column Values (Mapping):

##In the gender column, map the values as follows:
▪ male and Male should be updated to M

▪ female and Female should be updated to F

```python
df.columns
```

```python
d={'male':'M','Male':'M',"female":'F','Female':'F'}
df['gender']=df['gender'].map(d)
df
```

#4. Remove Unwanted Characters:

##Remove any leading or trailing spaces in the name column.

```python
df['name']=df['name'].str.strip()
df
```

#5. Change Data Type:

##Convert the age column to integer type, and the salary column to int type.

```python
# df['age']=df['age'].fillna(0)
df['age']=df['age'].astype(int)
df['salary']=df['salary'].astype(int)
df
```

#6. Replace Specific Values:

##In the employment_status column, replace missing values with 'contract'.

```python
df['employment_status']=df['employment_status'].fillna('contract')
df
```

#7. Handle Missing Values:

##o Fill missing values in the age column with the average age of the employees.
##o Fill missing values in the salary column with the median salary.
##o For missing join_date, use forward fill (ffill).

```python

avg=df['age'].mean()
avg1=df['salary'].median()

df['age']=df['age'].fillna(avg)
df['salary']=df['salary'].fillna(avg1)
df['join_date']=df['join_date'].fillna(method='ffill')


df
```

#8. Remove Duplicates:
o Ensure there are no duplicate rows in the dataset.

```python
df.drop_duplicates(inplace=True)
df
```

#9. Apply String Function:
o Capitalize the first letter of each word in the department column.

```python
df['department']=df['department'].str.capitalize()
df
```

#10. Drop Unwanted Columns:
o Remove the exp_years column from the dataset.

```python
df.drop('exp_years',axis=1,inplace=True)
df
```

#11. Remove Specific Rows:
o Drop rows where salary is above 80,000 to focus on mid-level employees.

```python
var=df[df['salary']> 80000]
var
df.drop(df.index[2],inplace=True)
df
```

#12. Reset Index:
o After cleaning, reset the index of the DataFrame.

```python
df.drop('index',axis=1,inplace=True)
df.reset_index(inplace=True)
df
```

#13. Set Column as Index:
o Set the employee_id as the index of the DataFrame.

```python
df.set_index('employee_id',inplace=True)
df
```
