```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/winequalityN.csv")
df.head()
```

```python
sns.jointplot(x='quality',y='alcohol',data=df,hue='type')
plt.show()
```

```python
sns.jointplot(x='quality',y='alcohol',data=df,kind='hex')
plt.show()
```

```python
sns.jointplot(x='residual sugar',y='alcohol',data=df,kind='hex')
```

```python
sns.jointplot(x='quality',y='alcohol',data=df,hue='quality')
```

```python
sns.jointplot(x='quality',y='chlorides',data=df,hue='chlorides')
plt.show()
```

```python
sns.jointplot(x='alcohol',y='chlorides',data=df,hue='chlorides')
plt.show()
```

```python
sns.jointplot(x='alcohol',y='chlorides',data=df,kind='hex')
plt.show()
```

```python
sns.jointplot(x='alcohol',y='chlorides',data=df,kind='kde')
plt.show()
```
