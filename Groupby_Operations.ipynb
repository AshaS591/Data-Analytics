#we can use morethan one column as a categorial column in groupy class

```python
var3=df.groupby(['department','salary'])
ds=var3.get_group(('IT','medium'))
ds
```

```python
var1=df.groupby(['department','left'])
var2=var1.get_group(('sales',1))['employee_id'].count()
print(var2)
```

```python
var2=var1.get_group(('IT',1))['employee_id'].count()
print(var2)
```

```python
var3=df.groupby(['department'])
var4=var3.get_group(('support'))['number_project'].sum()
var4
```

```python
var5=df.groupby(['promotion_last_5years','department'])
var6=var5.get_group((1,'accounting'))['employee_id'].count()
var6
```

```python
r_d=var5.get_group((1,'RandD'))['employee_id'].count()
var6+r_d
```

```python
var7=df.groupby(['salary','department'])
low=var7.get_group(('low','marketing'))['employee_id'].count()
high=var7.get_group(('high','marketing'))['employee_id'].count()
low+high
```
