```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/amazon_fires.csv', encoding='ISO-8859-1')
```

#1. Renaming the columns

```python
df.columns
```

```python
d={'ano':'year', 'mes':'month', 'estado':'state', 'numero':'no_of_fires', 'encontro':'date'}
df.rename(columns = d,inplace=True)
```

```python
df
```

#2. Removing duplicates rows

```python
df.drop_duplicates(inplace=True)
df
```

#3. Changing the data present in a particular column

```python
df.info()
```

```python
df['month'].unique()
```

```python
d={'Janeiro':'jan', 'Fevereiro':'feb', 'Março':'march', 'Abril':'april', 'Maio':'may', 'Junho':'june', 'Julho':'july',
       'Agosto':'august', 'Setembro':'september', 'Outubro':'october', 'Novembro':'november', 'Dezembro':'december'}
df['month']=df['month'].map(d)
df
```

#4. Removing unwanted data present in a column

```python
df['no_of_fires']=df['no_of_fires'].str.strip(' Fires')
df
```

# 5. Applying inbuilt functions on a particular column

```python
df['month']=df['month'].str.title()
df
```

# 9. Changing a particular data

```python
df['month']=df['month'].str.replace('d','D')
df
```
