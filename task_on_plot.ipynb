```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/sales_data_boolean_filtering.csv')
```

```python
df
```

```python
df.info()
```

```python
df['OrderDate']=pd.to_datetime(df['OrderDate'])
```

```python
df
```

```python
df['OrderDate'].unique()
```

```python
df['year']=[y.year for y in df['OrderDate']]
```

```python
var=df.groupby('year')['TotalSales'].sum()
df1=pd.DataFrame(var)
plt.plot(df1.index,df1['TotalSales'],marker='o',color='blue',markerfacecolor='orange',markeredgecolor='orange')
plt.title('Total Sales over year',color='red',fontsize=15)
plt.xlabel('Year',color='green',fontsize=15)
plt.ylabel('Total Sales',color='green',fontsize=15)
plt.show()
```

```python
var=df.groupby('Category')['TotalSales'].sum()
df1=pd.DataFrame(var)
plt.plot(df1.index,df1['TotalSales'],marker='o',color='blue',markerfacecolor='orange',markeredgecolor='orange')
plt.title('Total Sales over category ',color='red',fontsize=15)
plt.xlabel('category ',color='green',fontsize=15)
plt.ylabel('Total Sales',color='green',fontsize=15)
plt.show()
```

```python
var=df.groupby('Product')['TotalSales'].sum()
var
df1=pd.DataFrame(var)
plt.plot(df1.index,df1['TotalSales'],marker='o',color='blue',markerfacecolor='orange',markeredgecolor='orange')
plt.title('Total Sales over product ',color='red',fontsize=15)
plt.xlabel('product ',color='green',fontsize=15)
plt.ylabel('Total Sales',color='green',fontsize=15)
plt.xticks(rotation=90)
plt.show()
# plt.tight_layout()
```

```python
var=df.groupby('Region')['TotalSales'].sum()
df1=pd.DataFrame(var)
plt.plot(df1.index,df1['TotalSales'],marker='o',color='blue',markerfacecolor='orange',markeredgecolor='orange')
plt.title('Total Sales over category ',color='red',fontsize=15)
plt.xlabel('Region ',color='green',fontsize=15)
plt.ylabel('Total Sales',color='green',fontsize=15)
plt.show()
```
