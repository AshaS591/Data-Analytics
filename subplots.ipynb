```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[2,3,4,5,6,9]
y=[3,5,8,2,9,4]
p=np.sin(x)
q=np.cos(y)

z=np.arange(-5,6)
a=z**2

c=np.arange(6,-7,-1)
b=c**2
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,facecolor='green')
fig.suptitle('Multiple Grids',color='yellow')
axes[0,0].plot(x,y,color='crimson',marker='*',markeredgecolor='black',label='asha')
axes[0,1].plot(p,q,color='navy',marker='d',markerfacecolor='white',markeredgecolor='black',label='Sine_Cos')
axes[1,1].plot(z,a,color='navy',marker='*',markerfacecolor='white',markeredgecolor='black',label='parabola')
axes[1,0].plot(b,c,color='orange',marker='*',markerfacecolor='yellow',markeredgecolor='navy',label='parabola')


axes[0,0].set_facecolor('pink')
fig.legend()
fig.tight_layout()
```

#Annotate

```python
x=np.arange(-5,6)
y=x**2
plt.plot(x,y,marker='o',markerfacecolor='orange',markeredgecolor='black',color='green')
plt.axhline(15,color='blue')
plt.axvline(0,color='red')

for index in range(len(x)):
  plt.annotate(f"{x[index],y[index]}",(x[index],y[index]),ha='left',color='brown')
```

```python
marks=[99,95,93,95,93]
subjects=['bio','physics','chemistry','kannada','maths']

plt.pie(marks,labels=subjects,shadow=True,autopct='%0.0f%%',explode=[0.1,0,0,0,0])
plt.legend(marks)
plt.show()
```
