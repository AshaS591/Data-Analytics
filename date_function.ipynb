```python
df=pd.read_csv("https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/amazon_fires.csv",encoding='ISO-8859-1')
df
```

```python
df['encontro']=pd.to_datetime(df['encontro'])
df
```

```python
df.info()
```

```python
df['day']=[day.day for day in df['encontro']]
df
```

```python
df['month']=[day.month for day in df['encontro']]
df
```

```python
df['year']=[day.year for day in df['encontro']]
df
```
