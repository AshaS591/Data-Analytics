```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/amazon_fires.csv',encoding='ISO-8859-1')
df
```

```python
new={'ano':'year','mes':'month','estado':'state','numero':'no_of_fires','encontro':'date'}
df.rename(columns=new,inplace='True')
df
```

```python
df['month'].unique()
```

```python
new={'Janeiro':'january', 'Fevereiro':'february', 'MarÃ§o':'march', 'Abril':'april', 'Maio':'may', 'Junho':'june', 'Julho':'july',
       'Agosto':'august', 'Setembro':'september', 'Outubro':'october', 'Novembro':'november', 'Dezembro':'december'}
df['month']=df['month'].map(new)
df
```

```python
df['no_of_fires']=df['no_of_fires'].str.strip(' Fires')
df
```

```python
df['no_of_fires']=df['no_of_fires'].astype(float)
```

```python
# avg=df['no_of_fires'].mean()
df['no_of_fires']=df['no_of_fires'].fillna(0)
```

```python
df['no_of_fires']=df['no_of_fires'].astype(int)
```

#1) Total number of fires in the month of January

```python
var=df.groupby(['month'])
jan=var.get_group('january')['no_of_fires'].sum()
jan
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/titanic.csv')
df.head()
```

#4) how many ppl survived in b5 cabin

```python
df['survived'].value_counts()
```

```python
survived=df[(df['survived']==1) & (df['cabin']=='B5')]['survived'].count()
survived
```

```python
var2=df.groupby(['survived','cabin'])
data=var2.get_group((1,'B5'))['survived'].count()
data
```

# 5) count of females whose age is less than 20 and survived

```python
age=df[(df['age']<20) &(df['survived']==1) & (df['sex']=='female')]['survived'].count()
age
```

#6) Total fare paid by embarked S females

```python
females=df.groupby(['embarked','sex'])
fare=females.get_group(('S','female'))['fare'].sum()
fare
```

```python
fare1=df[(df['embarked']=="S") & (df['sex']=='female')]['fare'].sum()
fare1
```

#7) count of ppl in pclass 3 having only 1 person with them

```python
count=df[(df['pclass']==3) & (df['sibsp']==1)]['pclass'].count()
count
```

```python
ppl=df.groupby(['pclass','sibsp'])
count_ppl=ppl.get_group((3,1))['pclass'].count()
count_ppl
```

#8) maximum fare paid in embarked Q

```python
max_fare=df.groupby('embarked')
max_Q=max_fare.get_group('Q')['fare'].max()
max_Q
```

#9) Avg time spend by the ppl in support department who got promoted

```python
df=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/hr_data.csv')
df.head(3)
```

```python
time_spend=df.groupby(['department','promotion_last_5years'])
avg_time=time_spend.get_group(('support',1))['time_spend_company'].mean()
avg_time
```

```python
time=df[(df['department']=='support') & (df['promotion_last_5years']==1)]['time_spend_company'].mean()
time
```
