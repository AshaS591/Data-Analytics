```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({'Name':['Asha','Vyshu','harshe','arya'],
                  'loc':['KA','AP','AP','KA'],
                 'id':[1,3,5,6]})
```

```python
df1
```

```python
df2=pd.DataFrame({'DName':['dev','research','sales','support'],
                  'loc':['KA','AP','AP','KA'],
                 })
df2
```

#Vertical Concatenation

```python
df3=pd.concat([df1,df2])
df3
```

#Horizontal Concatenation

```python
df4=pd.concat([df1,df2],axis=1)
df4
```
