```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python

plt.pie([90,99,95,98],labels=['maths','bio','chemistry','physics'],autopct='%0.0f%%',explode=[0,0.1,0,0],shadow=True)
plt.show()
```

```python
x=[90,45,67,23,56,76]
y=['sec1','sec2','sec3','sec4','sec5','sec6']
plt.pie(x,labels=y,autopct='%0.0f%%')
plt.show()
```
